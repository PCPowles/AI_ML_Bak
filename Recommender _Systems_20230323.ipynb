```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']
# "C:\Users\pcpow\OneDrive\Desktop\DataScience_Udemy_20230321\u.data"
ratings = pd.read_csv('C:/Users/pcpow/OneDrive/Desktop/DataScience_Udemy_20230321/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
# "C:\Users\pcpow\OneDrive\Desktop\DataScience_Udemy_20230321\u.item"
movies = pd.read_csv('C:/Users/pcpow/OneDrive/Desktop/DataScience_Udemy_20230321/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings1 = pd.merge(movies, ratings)
```

```python
movieRatings = ratings1.pivot_table(index=['user_id'], columns=['title'], values='rating')
movieRatings.head()
```

```python
starWarsRatings = movieRatings['Star Wars (1977)']
starWarsRatings.head()
```

```python
starWarsRatings = movieRatings['101 Dalmatians (1996)']
starWarsRatings.head()
```

```python
#  corrwith function

similarMovies = movieRatings.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)


```

```python
# Improving the results of movie similarities

import numpy as np
movieStats = ratings1.groupby('title').agg({'rating': [np.size, np.mean]})
#movieStats.head(25)
movieStats.head()
```

```python
popularMovies = movieStats['rating']['size']>=100
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
df.head
```

```python
df.sort_values(['similarity'], ascending=False)[:15]
```
