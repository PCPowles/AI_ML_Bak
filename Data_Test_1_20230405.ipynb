```python
from pandas import read_csv
from pandas import DataFrame
from pandas import Grouper
from matplotlib import pyplot
# series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
# "C:\Users\pcpow\OneDrive\Desktop\ML-Poject_20230404\daily-minimum-temperatures.csv"
series = read_csv('C:/Users/pcpow/OneDrive/Desktop/ML-Poject_20230404/daily-minimum-temperatures.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
groups = series.groupby(Grouper(freq='A'))
years = DataFrame()
for name, group in groups:
 years[name.year] = group.values
years.plot(subplots=True, legend=False)
pyplot.show()
```
