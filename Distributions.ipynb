```python
#The exponential probability distribution or Power law
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

```python
#Binomial probability mass function
from scipy.stats import binom
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
#plt.plot(x, binom.pmf(x))
#plt.plot(x, binom.pmf(x, 10, 0.5))
#plt.plot(x, binom.pmf(x, 10, 0.75))
plt.plot(x, binom.pmf(x, 10, 0.3))
```

```python
#Poisson probability mass function
#With average of 500 visitors per day, what's the odds of
seeing 550 visitors on a given day?
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 500
x = np.arange(400, 600, 0.50)
plt.plot(x, poisson.pmf(x, mu))


#The odds of seeing 550 visitors on a given day, it turns out, 
#comes out to about 0.002 or 0.2% probability.

# .pdf = Probability Density Function
# .pmf = Probability Mass Function

```
